```python
from kashgari.corpus import ChinaPeoplesDailyNerCorpus
from kashgari.embeddings import BERTEmbedding
import numpy as np
```

```python
import os
# set gpu
os.environ["CUDA_VISIBLE_DEVICES"] = "2"
```

```python
embedding = BERTEmbedding('bert-base-chinese', 100) #load word embedding
```

```python
#load training data
train_x = np.load('Data/data.npy')
train_y = np.load('Data/label.npy')
```

```python
from keras.callbacks import ModelCheckpoint
from keras.callbacks import TensorBoard
```

```python
from kashgari.tasks.seq_labeling import BLSTMCRFModel
model = BLSTMCRFModel(embedding)#load model
```

```python
# set TesorBoard and training
tf_board_callback = TensorBoard(log_dir='logs', update_freq=100)
model.fit(train_x, train_y, batch_size=32, epochs=10 ,fit_kwargs={'callbacks': [tf_board_callback]})
```

```python
model.save('Model')#save model
```
